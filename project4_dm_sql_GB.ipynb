## Gradient Boosting

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

##### fit a gradient boosting classifier with default parameters to get a baseline idea of the performance


```python
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
# > GradientBoostingClassifier(criterion=â€™friedman_mseâ€™, init=None,
#  learning_rate=0.1, loss=â€™devianceâ€™, max_depth=3,
#  max_features=None, max_leaf_nodes=None,
#  min_impurity_split=1e-07, min_samples_leaf=1,
#  min_samples_split=2, min_weight_fraction_leaf=0.0,
#  n_estimators=100, presort=â€™autoâ€™, random_state=None,
#  subsample=1.0, verbose=0, warm_start=False)
y_pred = gb.predict(X_test)
print(f'Training Score: {gb.score(X_train, y_train)}')
print(f'Testing Score: {gb.score(X_test, y_test)}')
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

##### Learing Rate


```python
learning_rates = [1, 0.5, 0.25, 0.1, 0.05, 0.01]
train_results = []
test_results = []
for eta in learning_rates:
   model = GradientBoostingClassifier(learning_rate=eta)
   model.fit(X_train, y_train)
   train_pred = model.predict(X_train)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   train_results.append(roc_auc)
   y_pred = model.predict(X_test)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   test_results.append(roc_auc)
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(learning_rates, train_results, color = 'Green', label='Train AUC')
line2, = plt.plot(learning_rates, test_results, color = 'Magenta', label='Test AUC')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel("AUC score", fontsize=15)
plt.xlabel("learning rate", fontsize=15)
plt.title('Optimizing Learning Rate',fontsize=15);
plt.show()
```

##### a high learning rate results in overfitting. For this data, a learning rate of 0.1 is optimal.

##### N_estimators

```python
n_estimators = [1, 2, 4, 8, 16, 32, 64, 100, 250, 500, 1000 ]
train_results = []
test_results = []
for estimator in n_estimators:
   model = GradientBoostingClassifier(n_estimators=estimator)
   model.fit(X_train, y_train)
   train_pred = model.predict(X_train)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   train_results.append(roc_auc)
   y_pred = model.predict(X_test)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   test_results.append(roc_auc)
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(n_estimators, train_results, color = 'Green', label='Train AUC')
line2, = plt.plot(n_estimators, test_results, color = 'Magenta', label='Test AUC')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel("AUC score", fontsize=15)
plt.xlabel("n_estimators", fontsize=15)
plt.title('Optimizing N Estimators',fontsize=15);
plt.show()
```

##### Increasing the number of estimators may result in overfitting also. In our case, using 250 trees is optimal.

##### max_depth

```python
max_depths = np.linspace(1, 32, 32, endpoint=True)
train_results = []
test_results = []
for max_depth in max_depths:
   model = GradientBoostingClassifier(max_depth=max_depth)
   model.fit(X_train, y_train)
   train_pred = model.predict(X_train)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   train_results.append(roc_auc)
   y_pred = model.predict(X_test)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   test_results.append(roc_auc)
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(max_depths, train_results, color = 'Green', label='Train AUC')
line2, = plt.plot(max_depths, test_results,  color = 'Magenta', label='Test AUC')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel("AUC score", fontsize=15)
plt.xlabel("Tree Depth", fontsize=15)
plt.title('Optimizing Max Depth',fontsize=15);
plt.show()
```

We see that our model overfits for large depth values. The tree perfectly predicts all of the train data, however, it fails to generalize the findings for new data

##### min_samples_split

```python
min_samples_splits = np.linspace(0.1, 1.0, 10, endpoint=True)
train_results = []
test_results = []
for min_samples_split in min_samples_splits:
   model = GradientBoostingClassifier(min_samples_split=min_samples_split)
   model.fit(X_train, y_train)
   train_pred = model.predict(X_train)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   train_results.append(roc_auc)
   y_pred = model.predict(X_test)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   test_results.append(roc_auc)
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(min_samples_splits, train_results, color = 'Green', label='Train AUC')
line2, = plt.plot(min_samples_splits, test_results,  color = 'Magenta', label='Test AUC')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel("AUC score", fontsize = 15)
plt.xlabel("min samples split", fontsize = 15)
plt.title('Optimizing Min Sample Split',fontsize=15);
plt.show()
```

##### We can clearly see that when we require all of the samples at each node, the model cannot learn enough about the data. This is an underfitting case.

##### min_samples_leaf

```python
min_samples_leafs = np.linspace(0.1, 0.5, 5, endpoint=True)
train_results = []
test_results = []
for min_samples_leaf in min_samples_leafs:
   model = GradientBoostingClassifier(min_samples_leaf=min_samples_leaf)
   model.fit(X_train, y_train)
   train_pred = model.predict(X_train)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   train_results.append(roc_auc)
   y_pred = model.predict(X_test)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   test_results.append(roc_auc)
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(min_samples_leafs, train_results, color = 'Green', label='Train AUC')
line2, = plt.plot(min_samples_leafs, test_results, color = 'Magenta', label='Test AUC')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel("AUC score", fontsize=15)
plt.xlabel("min samples leaf",fontsize=15)
plt.title('Optimizing Min Sample Leaf',fontsize=15);
plt.show()
```

##### 0.5 is Optimal.

##### max_features

```python
max_features = list(range(1,X_train.shape[1]))
train_results = []
test_results = []
for max_feature in max_features:
   model = GradientBoostingClassifier(max_features=max_feature)
   model.fit(X_train, y_train)
   train_pred = model.predict(X_train)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   train_results.append(roc_auc)
   y_pred = model.predict(X_test)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   test_results.append(roc_auc)
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(max_features, train_results, color = 'Green', label='Train AUC')
line2, = plt.plot(max_features, test_results, color = 'Magenta', label='Test AUC')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel("AUC score",fontsize=15)
plt.xlabel("max features",fontsize=15)
plt.title('Optimizing Max Features',fontsize=15);
plt.show()
```

##### Increasing max_features to consider all of the features results in an overfitting in this case. Using max_features = 20 seems to get us the optimal performance.
